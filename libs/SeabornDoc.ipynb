## Seaborn é uma biblioteca de visualização de dados para Python que se baseia em cima da biblioteca Matplotlib. Seu objetivo é tornar a criação de gráficos estatísticos mais fácil, permitindo que os usuários se concentrem na análise de dados em vez de nos detalhes da criação de gráficos.

### Instalação:

```python
pip install seaborn
```

### Importação:
#### Para usar Seaborn em seu código Python, você precisa primeiro importá-lo. O método mais comum de importação é o seguinte:

```python
import seaborn as sns
```

### Exemplos de uso:
#### Seaborn é capaz de criar uma variedade de gráficos estatísticos, incluindo gráficos de dispersão, gráficos de barras, gráficos de linha, histogramas e muito mais.

#### Para criar um gráfico de dispersão com uma linha de regressão usando Seaborn, você pode usar o seguinte código:

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.regplot(x="total_bill", y="tip", data=tips)

plt.show()
```

#### Neste exemplo, o conjunto de dados 'tips' é carregado usando o método load_dataset() fornecido pelo Seaborn. Em seguida, regplot() é usado para criar um gráfico de dispersão com uma linha de regressão. Finalmente, o gráfico é exibido usando o método show() do Matplotlib.

### Outro exemplo é a criação de um histograma usando Seaborn:



```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")

sns.histplot(data=iris, x="sepal_length", hue="species", kde=True)

plt.show()
```

#### Neste exemplo, o conjunto de dados 'iris' é carregado usando o método load_dataset() fornecido pelo Seaborn. Em seguida, histplot() é usado para criar um histograma com uma curva de densidade estimada para cada espécie de íris. Finalmente, o gráfico é exibido usando o método show() do Matplotlib.
